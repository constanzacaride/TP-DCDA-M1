Pregunta nro 11 -> Realizar un análisis RFM Base/s de datos a utilizar: Tabla Preg 5
#Un Análisis RFM es un modelo que permite segmentar clientes basado en 3 criterios
#Recency (R): Cuando fue la última vez que el cliente hizo una compra?
#Frequency (F): Con que frecuencia ha comprado dentro de un periodo de tiempo determinado
#Monetary (M): La cantidad de dinero que ese cliente ha gastado.

```python
#En este caso haré un gráfico (scaterplot) donde consideraré el valor del ticket promedio, la fecencia de compra y la última compra en el año 2017 (que es el que tengo completo) para cada cliente
#Genero nuevas columnas y completo con el año y dia juliano obtenido de la columna order_purchase_timestamp (ya en dt)
bbddPreg5_base['Año'] = bbddPreg5_base['order_purchase_timestamp'].dt.strftime('%Y')
bbddPreg5_base['diaJ'] = bbddPreg5_base['order_purchase_timestamp'].dt.strftime('%j')
bbddPreg5_base.sample(6) #Check
```

```python
#Hago un subset solo con los valores del Año 2017
Preg11=bbddPreg5_base[bbddPreg5_base.Año == "2017"]
Preg11.sample(10) #Check
```

```python
#Observo los max y min de las columnas a utilizar para ver los rangos de los valores y luego chequear que el gráfico coincida
print("DiaJ min-max:", Preg11["diaJ"].min(), "-", Preg11["diaJ"].max())
print("Cantidad de order_id:", Preg11["order_id"].count())
print("payment_value min-max:", Preg11["payment_value"].min(), "-", Preg11["payment_value"].max())
```

```python
#A partir de esta BBDD calculo, el ticket promedio de cada cliente, cuento la cantidad de veces que compró en 2017 y veo que dia del año fue la última compra
Preg11_info = pd.pivot_table(Preg11, values=['diaJ', 'order_id', 'payment_value'], index=['customer_unique_id'], aggfunc={'diaJ': np.max, 'order_id': np.size, 'payment_value': np.mean })
Preg11_info2 = Preg11_info.rename(columns={'diaJ': 'Dia última compra', 'order_id': 'Freciencia de compra', 'payment_value': 'Ticket promedio'})
print(Preg11_info2.info())
Preg11_info2.head()
```

```python
#Observo los max y min de las columnas a utilizar para ver los rangos de los valores y luego chequear que el gráfico coincida
print("Dia última compra min-max:", Preg11_info2["Dia última compra"].min(), "-", Preg11_info2["Dia última compra"].max())
print("Freciencia de compra min-max:", Preg11_info2["Freciencia de compra"].min(), "-", Preg11_info2["Freciencia de compra"].max())
print("Ticket promedio min-max:", Preg11_info2["Ticket promedio"].min(), "-", Preg11_info2["Ticket promedio"].max())
```

```python
#Hago el gáfico para evaluar el comportamiento de cada cliente con respecto a las tres variables
Preg11_info2.plot.scatter(x='Dia última compra',y='Freciencia de compra',c='Ticket promedio',figsize=(10, 6),colormap='viridis').set_title('Distribución del ticket-prom en fc de la última compra y la frecuencia')
print("Pregunta nro 11 -> Realizar un análisis RFM - RESULTADO")
plt.show()
```

```python
#1- Ranking de vendedores por 1-cantidad de items vendidos y 2-ticket promedio 
#Base/s de Datos a utilizar: Orders + Items + Sellers + Payments
Preg13s1 = pd.merge(Items, Sellers, on='seller_id', how='inner')
print('Filas-Columnas s1: ',Preg13s1.shape)
Preg13s2 = pd.merge(Preg13s1, Orders, on='order_id', how='inner')
print('Filas-Columnas s2: ',Preg13s2.shape)
Preg13s3 = pd.merge(Preg13s2, Payments, on='order_id', how='inner')
print('Filas-Columnas s3: ',Preg13s3.shape)
Preg13s3.head()
```

```python
# Controlo si hay o no valores nulos
print("Existen datos faltantes?",Preg13s3.isnull().any().any())
```

```python
#Veo donde están
print("Valores faltantes x columna")
Preg13s3.isnull().sum()
```

```python
# No me afectan esas columnas a mi análisis asi que no hago nada
# Cuento la cantidad de items que vendiò cada vendedor y el valor de ticket promedio
Preg13s_calc = pd.pivot_table(Preg13s3, values=['order_item_id', 'price'], index=['seller_id'], aggfunc={'order_item_id': np.sum, 'price': np.mean})
Preg13s_calc_rn = Preg13s_calc.rename(columns={'seller_id': 'Vendedor', 'order_item_id': 'Items-vendidos', 'price': 'Ticket-prom'})
print('Filas-Columnas s3: ',Preg13s_calc_rn.shape)
```

```python
print("Ranking por Items-vendidos")
Preg13s_calc_rn.sort_values(by='Items-vendidos', ascending=False).head().round(2)
```

```python
print("Ranking por Ticket-prom")
Preg13s_calc_rn.sort_values(by='Ticket-prom', ascending=False).head().round(2)
```

```python
#2- Seleccionar a los vendedores que estén en el cuartil superior en "ambas" categorias del punto anterior y darles un baneficio
#Calculo los cuartiles 
q3i, q2i, q1i = np.percentile(Preg13s_calc_rn["Items-vendidos"], [75, 50, 25])
q3t, q2t, q1t = np.percentile(Preg13s_calc_rn["Ticket-prom"], [75, 50, 25])
print("Cuartiles 25 50 75 items =", q1i,"-", q2i,"-", q3i)
print("Cuartiles 25 50 75 tickets =", q1t,"-", q2t,"-", q3t)
```

```python
#Elegimos solo a los vendedores que cumplen con ambos requicitos
Beneficio = Preg13s_calc_rn.loc[(Preg13s_calc_rn['Items-vendidos'] > q3i) & (Preg13s_calc_rn['Ticket-prom'] > q3t)]
print('Filas-Columnas: ',Beneficio.shape)
Beneficio.head()
```

```python
#143 de 3095 son un 4,6%
#Chequeamos que haya funcionado bien la selección
print("Items-vendidos min-max:", Beneficio["Items-vendidos"].min(), "-", Beneficio["Items-vendidos"].max())
print("Ticket-prom min-max:", Beneficio["Ticket-prom"].min(), "-", Beneficio["Ticket-prom"].max())
```

```python
#3- Hacer una mapa de calor de los dias/horas de mayores venta para hacer promos
#Base de Datos a utilizar -> bbddPreg5_base
#Traigo los columnas que mecesito y calculo dia de la semana y rango de hora
Preg13a = bbddPreg5_base[['order_id', 'order_purchase_timestamp']]
Preg13a['Dia-semana'] = Preg13a['order_purchase_timestamp'].dt.day_name(locale='Spanish')
Preg13a['Rango-hr'] = Preg13a['order_purchase_timestamp'].dt.hour
print('Filas-Columnas: ',Preg13a.shape)
Preg13a.head()
```

```python
OrFil = {'Lunes': 0, 'Martes': 1, 'Miércoles': 2, 'Jueves': 3, 'Viernes': 4, 'Sábado': 5, 'Domingo': 6}
Preg13_salida = Preg13a.pivot_table(index='Dia-semana', columns='Rango-hr', values='order_id', aggfunc='count', fill_value=0).reindex(OrFil)
Preg13_salida.head()
```

```python
plt.figure(figsize=(25, 8))
sns.heatmap(Preg13_salida, cmap="RdYlGn", linewidths = 0.75, linecolor = "white").set_title('Ventas por día-hora')
plt.show()
```

Pregunta nro 14 -> Imputar datos ausentes Base/s de datos a utilizar: Products

```python
#Realizo un ejemplo de como Imputar datos ausentes categóricos (con la media) y cuantitativos (con la moda)
#También como reemplazar los NaN con un dato arbitrario
#Anteriormente vi que en la tabla de Products las columnas product_category_name, product_name_lenght y product_photos_qty tenían NaN
#Para el ejemplo entonces traigo solo esas columnas de la tabla 
Preg14 = Products[['product_id', 'product_category_name', 'product_name_lenght', 'product_photos_qty']]
print(Preg14.dtypes)
Preg14.head()
```

```python
# Controlo si hay o no valores nulos
print("Existen datos faltantes?",Preg14.isnull().any().any())
```

```python
#Veo donde están
print("Valores faltantes x columna")
Preg14.isnull().sum()
```

```python
#Voy a hacer un groupby del primer reemplazo para chequear luego que funciona bien
Check1 = Preg14.groupby('product_category_name')['product_id'].count() 
print("numero de filas: ",Preg14.shape[0])
Check1
```

```python
#La categoría "cama_mesa_banho" es la que más se repite (la moda)
#Reemplazo los faltantes de la columna product_category_name (dato categorico) con la moda
Preg14['product_category_name'].fillna((Preg14['product_category_name'].mode().iloc[0]), inplace=True)
```

```python
#Chequeo - La frecuencia de la categoría "cama_mesa_banho" debe haber aumentado en 610 unidades 
Check2 = Preg14.groupby('product_category_name')['product_id'].count() 
print("numero de filas: ",Preg14.shape[0])
Check2
```

```python
#Reemplazo los faltantes de la columna product_name_lenght (dato cuantitativo) con la media
Preg14['product_name_lenght'].fillna((Preg14['product_name_lenght'].mean()), inplace=True)
```

```python
#Reemplazo los faltantes de la columna product_name_lenght (dato cuantitativo) con el valor 980
Preg14['product_photos_qty'].fillna(980, inplace=True)
```

```python
#Vuelvo a preguntar si aún hay valores faltantes
print("Existen datos faltantes?",Preg14.isnull().any().any())
```
